#### Problema 1
La suma de un número más 9 es igual a 21Paso 1: Explicación y resolución algebraicaPlanteamiento: Llamaremos $x$ al número desconocido. El enunciado dice que "un número ($x$) más nueve ($+ 9$) es igual ($=$) a 21".Ecuación:$$x + 9 = 21$$Despeje: Para dejar a la $x$ sola, el 9 que está sumando pasa al otro lado de la igualdad restando:$$x = 21 - 9$$Solución:$$x = 12$$

```python
# Problema 1: Hallar el número desconocido
# Ecuación: x + 9 = 21
resultado_total =int(input("Ingrese el resultado total: "))
numero_conocido = int(input("Ingrese el número conocido: "))

resultado_total = 21
numero_conocido = 9

# Despejamos x: x = 21 - 9
x = resultado_total - numero_conocido

print(f"El número buscado en el Problema 1 es: {x}")
```
