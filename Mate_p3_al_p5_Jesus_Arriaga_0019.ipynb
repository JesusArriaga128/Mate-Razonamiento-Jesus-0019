#### Problema 1
La resta inversaProblema: Un número al que se le resta 7 da como resultado 18. ¿Cuál es ese número?Paso 1: Explicación y resoluciónPlanteamiento: Sea $x$ el número. Al número ($x$) le restamos ($-$) siete ($7$) y obtenemos ($=$) dieciocho ($18$).Ecuación:$$x - 7 = 18$$Despeje: Para dejar la $x$ sola, el 7 que está restando pasa al otro lado de la igualdad con la operación opuesta: la suma.$$x = 18 + 7$$Solución:$$x = 25$$

```python
# Problema 3: El número al que se le resta 7
#resultado = 18
#resta = 7
resultado=int(input("Ingrese el resultado: "))
resta=int(input("Ingrese la resta: "))
# Despeje: x = 18 + 7
numero = resultado + resta
print(f"El número buscado es: {numero}")
```

#### problema 2
Ecuación de dos pasos (Triple más constante)Problema: El triple de un número más cuatro es igual a 31. ¿Cuál es ese número?Paso 1: Explicación y resoluciónPlanteamiento: * El triple de un número: $3x$Más cuatro: $+ 4$Es igual a: $= 31$Ecuación:$$3x + 4 = 31$$Despeje: * Primero quitamos lo que suma: el 4 pasa restando.$$3x = 31 - 4 \implies 3x = 27$$Luego quitamos lo que multiplica: el 3 pasa dividiendo.$$x = \frac{27}{3}$$

```python
# Problema 4: El triple de un número más 4
#total = 31
#suma = 4
multiplo = 3
total=int(input("Ingrese el total: "))
suma=int(input("Ingrese la suma: "))

# Paso 1: Restar el 4 -> 31 - 4 = 27
# Paso 2: Dividir entre 3 -> 27 / 3
x = (total - suma) / multiplo
print(f"El número buscado es: {int(x)}")
```

## Problema 5:
### Perímetro de un terreno (Rectángulo)
#### Enunciado: Un terreno tiene un largo que mide 5 metros más que el ancho. El perímetro total es de 50 metros.
#### Paso 1: Análisis y Solución
#### Planteamiento:Ancho: $x$Largo: $x + 5$ (mide 5 metros más que el ancho).Perímetro: Es la suma de los cuatro lados: $Largo + Largo + Ancho + Ancho = 50$.Ecuación:$$2(x + 5) + 2(x) = 50$$Despeje:Simplificamos los paréntesis:$$2x + 10 + 2x = 50$$
#### Agrupamos las $x$:$$4x + 10 = 50$$
#### Pasamos el 10 restando:$$4x = 50 - 10$$$$4x = 40$$Pasamos el 4 dividiendo:$$x = \frac{40}{4}$$
#### Solución: Ancho ($x$): 10 metros.Largo ($x + 5$): 15 metros.
#### Comprobación: $15 + 15 + 10 + 10 = 50$.

```python
# Problema 5: Perímetro de un terreno
# perimetro_total = 50
# diferencia_largo = 5
perimetro_total = int(input("Ingrese el perímetro total: "))
diferencia_largo = int(input("Ingrese la diferencia entre el largo y el ancho: "))

# Ecuación: 4x + 2*(diferencia) = perimetro
# Despejamos x (ancho): x = (perimetro - 2*diferencia) / 4
ancho = (perimetro_total - (2 * diferencia_largo)) / 4
largo = ancho + diferencia_largo

print(f"Las dimensiones son: Ancho = {int(ancho)}m, Largo = {int(largo)}m")
```
